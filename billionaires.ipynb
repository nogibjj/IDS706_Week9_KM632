# EDA


```python
df = pd.read_csv('/world_billionaires.csv')
gender_counts = df['gender'].value_counts().to_dict()
print("\nCount of individuals by gender:")
for gender, count in gender_counts.items():
    print(f"{gender} {count}", end=" ")
```

We see that the vast majority of the billionares are man. Below is a pie chart further illustrating the gap.

```python
gender_counts = df['gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Gender Distribution")
```

Interestingly, we see that that the majority of men are self made whereas majority of the women are NOT self-made.





```python
self_made_counts = df.groupby(['gender', 'selfMade']).size().unstack(fill_value=0)

# Plot the bar chart with counts
ax = self_made_counts.plot(kind='bar', stacked=True)
plt.title("Self-Made Status by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Self-Made", labels=["Not Self-Made", "Self-Made"])

# Add counts as text above each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Show the bar chart
plt.show()

```

```python
self_made_df = df[df['selfMade']]
self_made_categories = self_made_df['category'].value_counts().head(5)
print("Top 5 Categories for Self-Made Billionaires:")
print(self_made_categories)
```
